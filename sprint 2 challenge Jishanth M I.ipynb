# Task 2

```python
#Perform bivariate and multivariate analysis on categorical columns (identified in Task 1) with the help of appropriate graphs/charts.


# Bivariate Analysis (Categorical vs. Categorical)
plt.figure(figsize=(10,6))
pd.crosstab(index=data['WorkClass'],columns=data['Marital_Status']).plot(kind='bar')
pd.crosstab(index=data['WorkClass'],columns=data['Gender']).plot(kind='bar')


# multivariate analysis
plt.figure(figsize=(10,6))
sb.barplot(x=data['Marital_Status'],y=data['Age'],hue=data['WorkClass'])
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(10,6))
sb.barplot(x=data['Education'],y=data['Age'],hue=data['WorkClass'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Perform bivariate and multivariate analysis on numarical columns (identified in Task 1) with the help of appropriate graphs/charts.

# Bivariate Analysis (numarical vs. numarical)
sb.lineplot(x=data['Education_Num'],y=data['Hours_Per_Week'])

# multivariate analysis
sb.pairplot(data,vars=['Age', 'Education_Num', 'Hours_Per_Week'],diag_kind='kde',kind='scatter')
```
