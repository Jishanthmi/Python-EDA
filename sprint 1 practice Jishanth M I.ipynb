```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"C:\Users\Jishanth M I\Downloads\DS1_C5_S1_EmployeeAttrition_Data_Practice.csv")
data
```

```python
# Task 1
# Display the basic information about the dataset, which includes column names, count of non-null records, and data type of the columns. Also, display the count of missing values for each column.


# Display column names, non-null counts, and data types
column_info = data.info()

# Display count of missing values for each column
missing_values = data.isnull().sum()

print(column_info)
print(missing_values)
```

```python
# Task 2
# Define a function that identifies the data type of columns as categorical or numerical.
#Separate the names of categorical columns and numerical columns and save them in two different lists.


data_cat=data.select_dtypes(include='object')
print(data_cat.columns)

data_con=data.select_dtypes(include='number')
print(data_con.columns)
```

```python
# Task 3
#Perform descriptive analysis of each categorical column (identified in Task 2) by defining a function to print the following information.
#The number of categories (unique value) present in a column
#The mode of a column
#The missing values of a column
#Perform the missing value treatment, if required



# Number of unique categories
num_categories = data_cat.nunique()
print("Number of Categories:", num_categories)
        
# Mode of each column
mode_values = data_cat.mode()
print("Mode:", mode_values)
        
# Count of missing values for a specific column
missing_count = data_cat.isnull().sum()
print("Missing Values:", missing_count)
        
```

```python
# Task 4

# Use the Attrition column
plt.figure(figsize=(10, 5))

# Bar chart for count of attrited and existing employees
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='Attrition')
plt.title('Attrition Count')

# Pie chart for percentage of attrited and existing employees
plt.subplot(1, 2, 2)
attrition_percentage = data['Attrition'].value_counts(normalize=True) * 100
plt.pie(attrition_percentage, labels=attrition_percentage.index, autopct='%.2f %%')
plt.title('Attrition Percentage')

plt.tight_layout()
plt.show()

# Use the BusinessTravel column
plt.figure(figsize=(10, 5))

# Bar chart for employee count by type of business travel
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='BusinessTravel')
plt.title('Business Travel Count')

# Pie chart for employee percentage by type of business travel
plt.subplot(1, 2, 2)
business_travel_percentage = data['BusinessTravel'].value_counts(normalize=True) * 100
plt.pie(business_travel_percentage, labels=business_travel_percentage.index, autopct='%.2f%%')
plt.title('Business Travel Percentage')

plt.tight_layout()
plt.show()

# Use the Department column
plt.figure(figsize=(12, 6))

# Bar chart for employee count by department
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='Department')
plt.title('Department-wise Employee Count')

# Pie chart for department-wise employee allocation percentage
plt.subplot(1, 2, 2)
department_percentage = data['Department'].value_counts(normalize=True) * 100
plt.pie(department_percentage, labels=department_percentage.index, autopct='%.2f%%')
plt.title('Department-wise Employee Allocation Percentage')

plt.tight_layout()
plt.show()


# Use the EducationField column
plt.figure(figsize=(12, 6))

# Bar chart for employee count by education field
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='EducationField')
plt.title('Education Field-wise Employee Count')

# Pie chart for education field-wise employee allocation percentage
plt.subplot(1, 2, 2)
education_field_percentage = data['EducationField'].value_counts(normalize=True) * 100
plt.pie(education_field_percentage, labels=education_field_percentage.index, autopct='%.2f%%')
plt.title('Education Field-wise Employee Allocation Percentage')

plt.tight_layout()
plt.show()

# Use the EnvironmentSatisfaction column
plt.figure(figsize=(12, 6))

# Bar chart for employee count by environment satisfaction score
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='EnvironmentSatisfaction')
plt.title('Employee Count by Environment Satisfaction')

# Display percentage on top of each bar
ax = sb.countplot(data=data, x='EnvironmentSatisfaction')
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() /data.shape[0]*100, 2)}%"         
                                                                    
    x = bar.get_x() + bar.get_width() /2                            
    y = bar.get_height()                                            
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" )
    
    
# Use the Gender column
plt.subplot(1, 2, 2)
gender_count = data['Gender'].value_counts()
plt.bar(gender_count.index, gender_count.values)
plt.title('Gender-wise Employee Count')

plt.tight_layout()
plt.show()

# Use the MaritalStatus column
plt.figure(figsize=(12, 6))

# Bar chart for employee count by marital status
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='MaritalStatus')
plt.title('Marital Status-wise Employee Count')

# Display percentage on top of each bar
ax = sb.countplot(data=data, x='MaritalStatus')
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() /data.shape[0]*100, 2)}%"         
                                                                    
    x = bar.get_x() + bar.get_width() /2                            
    y = bar.get_height()                                            
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" )

# Use the OverTime column
plt.subplot(1, 2, 2)
overtime_percentage = data['OverTime'].value_counts(normalize=True) * 100
plt.pie(overtime_percentage, labels=overtime_percentage.index, autopct='%.2f%%')
plt.title('Percentage of Employees Working Overtime')

plt.tight_layout()
plt.show()

# Use the PerformanceRating column
plt.figure(figsize=(12, 6))

# Pie chart for performance rating-wise employee percentage
plt.subplot(1, 2, 1)
performance_rating_percentage = data['PerformanceRating'].value_counts(normalize=True) * 100
plt.pie(performance_rating_percentage, labels=performance_rating_percentage.index, autopct='%.2f%%')
plt.title('Performance Rating-wise Employee Percentage')

# Use the 'WorkLifeBalance' column
plt.subplot(1, 2, 2)
worklife_balance_percentage = data['WorkLifeBalance'].value_counts(normalize=True) * 100
plt.pie(worklife_balance_percentage, labels=worklife_balance_percentage.index, autopct='%.2f%%')
plt.title('Work-Life Balance-wise Employee Percentage')

plt.tight_layout()
plt.show()
```

```python
# Task 5
#Perform descriptive analysis of each numerical column by defining a function to print the following information.
#The measures of central tendency (mean, median)
#The standard deviation of a column 
#The missing values of a column
#The skewness of a column



# Measures of central tendency
mean_value = data_con.mean()
median_value = data_con.median()
print("Mean:", mean_value)
print("Median:", median_value)
        
# Standard deviation
std_deviation = data_con.std()
print("Standard Deviation:", std_deviation)

# Count of missing values
missing_count = data_con.isnull().sum()
print("Missing Values:", missing_count)

# Skewness
from scipy.stats import skew
skewness = skew(data_con.dropna())  # Drop missing values before calculating skewness
print("Skewness:", skewness)
```
