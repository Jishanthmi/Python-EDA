```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"C:\Users\Jishanth M I\Downloads\DS1_C5_S2_EmployeeAttrition_Data_Practice.csv")
data
```

# Task 1

```python
# Analyze business travel category-wise employee attrition using a count plot
plt.figure(figsize=(8, 6))
sb.countplot(data=data, x="BusinessTravel", hue="Attrition")
plt.title("Business Travel vs. Employee Attrition")
plt.xlabel("Business Travel")
plt.ylabel("Count")
plt.show()

# Analyze education field-wise employee attrition using a bar chart
plt.figure(figsize=(10, 6))
sb.countplot(data=data, x="EducationField", hue="Attrition")
plt.title("Education Field vs. Employee Attrition")
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Analyze department-wise employee attrition using a bar chart
plt.figure(figsize=(8, 6))
sb.countplot(data=data, x="Department", hue="Attrition")
plt.title("Department vs. Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()

# Analyze environment satisfaction score-wise employee attrition using a bar chart
plt.figure(figsize=(8, 6))
sb.countplot(data=data, x="EnvironmentSatisfaction", hue="Attrition")
plt.title("Environment Satisfaction vs. Employee Attrition")
plt.xlabel("Environment Satisfaction")
plt.ylabel("Count")
plt.show()

# Analyze gender-wise employee attrition using a count plot
plt.figure(figsize=(6, 6))
sb.countplot(data=data, x="Gender", hue="Attrition")
plt.title("Gender vs. Employee Attrition")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Analyze marital status-wise employee attrition using a count plot
plt.figure(figsize=(6, 6))
sb.countplot(data=data, x="MaritalStatus", hue="Attrition")
plt.title("Marital Status vs. Employee Attrition")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()

# Analyze overtime category-wise employee attrition using a pie chart
attrition_by_overtime = data.groupby("OverTime")["Attrition"].value_counts()
attrition_by_overtime.plot(kind="pie", autopct="%.2f%%", figsize=(10, 6))
plt.title("Attrition by Overtime")
plt.show()

# Analyze performance rating-wise employee attrition using a count plot
plt.figure(figsize=(6, 6))
sb.countplot(data=data, x="PerformanceRating", hue="Attrition")
plt.title("Performance Rating vs. Employee Attrition")
plt.xlabel("Performance Rating")
plt.ylabel("Count")
plt.show()

# Analyze job satisfaction category-wise employee attrition using a count plot
plt.figure(figsize=(8, 6))
sb.countplot(data=data, x="JobSatisfaction", hue="Attrition")
plt.title("Job Satisfaction vs. Employee Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")
plt.show()
```

```python
#Perform the bi-variate analysis between categorical and numerical variables using an appropriate graph/chart.
#Use the 'Age' column to display age group-wise distribution of the employee attrition using a histogram and a box plot.
#Use the 'HourlyRate' column to display the hourly salary-wise employee attrition using a histogram and a box plot.
#Use the 'Monthly Income' column to display the monthly salary-wise employee attrition using a histogram and a box plot.
#Use the 'Distance From Home' column to display the employee attrition by distance from home (in kilometers) using a histogram and a box plot.
#Use the 'TotalWorkingYears' column to display the tenure-wise employee attrition using a histogram and a box plot.



# Analyze age group-wise distribution of employee attrition using a histogram and a box plot
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sb.histplot(data=data, x="Age", hue="Attrition", multiple="stack", bins=20)
plt.title("Age Group-Wise Distribution of Employee Attrition (Histogram)")
plt.xlabel("Age")
plt.ylabel("Count")

# Box Plot
plt.subplot(1, 2, 2)
sb.boxplot(data=data, x="Attrition", y="Age")
plt.title("Age Group-Wise Distribution of Employee Attrition (Box Plot)")
plt.xlabel("Attrition")
plt.ylabel("Age")

plt.show()

# Analyze hourly salary-wise employee attrition using a histogram and a box plot
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sb.histplot(data=data, x="HourlyRate", hue="Attrition", multiple="stack", bins=20)
plt.title("Hourly Salary-Wise Employee Attrition (Histogram)")
plt.xlabel("Hourly Rate")
plt.ylabel("Count")

# Box Plot
plt.subplot(1, 2, 2)
sb.boxplot(data=data, x="Attrition", y="HourlyRate")
plt.title("Hourly Salary-Wise Employee Attrition (Box Plot)")
plt.xlabel("Attrition")
plt.ylabel("Hourly Rate")

plt.show()

# Analyze monthly salary-wise employee attrition using a histogram and a box plot
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sb.histplot(data=data, x="MonthlyIncome", hue="Attrition", multiple="stack", bins=20)
plt.title("Monthly Salary-Wise Employee Attrition (Histogram)")
plt.xlabel("Monthly Income")
plt.ylabel("Count")

# Box Plot
plt.subplot(1, 2, 2)
sb.boxplot(data=data, x="Attrition", y="MonthlyIncome")
plt.title("Monthly Salary-Wise Employee Attrition (Box Plot)")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")

plt.show()

# Analyze employee attrition by distance from home using a histogram and a box plot
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sb.histplot(data=data, x="DistanceFromHome", hue="Attrition", multiple="stack", bins=20)
plt.title("Employee Attrition by Distance From Home (Histogram)")
plt.xlabel("Distance From Home (km)")
plt.ylabel("Count")

# Box Plot
plt.subplot(1, 2, 2)
sb.boxplot(data=data, x="Attrition", y="DistanceFromHome")
plt.title("Employee Attrition by Distance From Home (Box Plot)")
plt.xlabel("Attrition")
plt.ylabel("Distance From Home (km)")

plt.show()

# Analyze tenure-wise employee attrition using a histogram and a box plot
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sb.histplot(data=data, x="TotalWorkingYears", hue="Attrition", multiple="stack", bins=20)
plt.title("Tenure-Wise Employee Attrition (Histogram)")
plt.xlabel("Total Working Years")
plt.ylabel("Count")

# Box Plot
plt.subplot(1, 2, 2)
sb.boxplot(data=data, x="Attrition", y="TotalWorkingYears")
plt.title("Tenure-Wise Employee Attrition (Box Plot)")
plt.xlabel("Attrition")
plt.ylabel("Total Working Years")

plt.show()
```

# Task 3

```python
#Visualize the job role-wise count of the attrited employees who were travelling rarely, using a horizontal bar chart.


# Filter the data for attrited employees who were traveling rarely
attrited_rarely = data[(data['Attrition'] == 'Yes') & (data['BusinessTravel'] == 'Travel_Rarely')]

# Count the attrited employees by job role
attrited_by_job_role = attrited_rarely['JobRole'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
attrited_by_job_role.plot(kind='bar', color='skyblue')
plt.title("Job Role-Wise Count of Attrited Employees (Travel Rarely)")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.show()
```

# Task 4

```python
#Visualize the job role-wise employee attrition from the Life Sciences and Medical education background using a horizontal bar chart


# Filter the data for employees with Life Sciences and Medical education background and attrition
life_sciences_medical = data[(data['EducationField'].isin(['Life Sciences', 'Medical'])) & (data['Attrition'] == 'Yes')]

# Count the attrited employees by job role
attrition_by_job_role = life_sciences_medical['JobRole'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
attrition_by_job_role.plot(kind='bar', color='skyblue')
plt.title("Job Role-Wise Employee Attrition (Life Sciences and Medical)")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.show()
```
